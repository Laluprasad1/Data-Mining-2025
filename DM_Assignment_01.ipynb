2. Handle missing values using mean, median, or mode imputation.

```python
df.isna()
```

```python
print("Mean of Age:", df['Age'].mean())
print("Median of Age:", df['Age'].median())
print("Mode of Age:", df['Age'].mode()[0])
```

```python
print("Mean of Salary:", df['Salary'].mean())
print("Median of Salary:", df['Salary'].median())
print("Mode of Salary:", df['Salary'].mode()[0])
```

```python
print("Mean of Job Satisfatcion:", df['Job_Satisfaction'].mean())
print("Median of Job Satisfatcion:", df['Job_Satisfaction'].median())
print("Mode of Job Satisfatcion:", df['Job_Satisfaction'].mode()[0])
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Salary'] =df['Salary'].fillna(df['Salary'].mean())
```

```python
df['Job_Satisfaction'] = df['Job_Satisfaction'].fillna(df['Job_Satisfaction'].mean())
```
