# **Dimensionality Reduction**
<br> Use PCA (Principal Components Analysis) to reduce dimensions of a dataset.
Visualize the reduced features.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

#Step - 1: Load the Iris dataset

```python
iris = load_iris()
x = iris.data    #feature
y = iris.target   #labels
```

```python
x
```

```python
x.shape   #size of the x
```

```python
y
```

```python
y.shape   #size of the y
```

#Step 2: Preprocess the data (standardization)

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
x_scaled.shape
```

#Step 3: Apply PCA to reduce to 2 dimensions

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
```

```python
x_pca.shape
```

```python
x_pca
```

# Visualize the reduced features (2D scatter plot)

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=y, cmap='viridis', edgecolors='k', s=100)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset (2 components)')
plt.colorbar(label='Target Class')
plt.show()
```

```python
#optional: Explained  Variance Ratio
print("Explained Variance Ratio of Each Component: ")
pca.explained_variance_ratio_
```
