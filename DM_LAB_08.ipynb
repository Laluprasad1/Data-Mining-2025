#**1. Data Transformation**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
#Example Data
data = {'Category': ['A', 'B', 'C', 'A', 'B']}
df = pd.DataFrame(data)
```

```python
df
```

```python
#Label Encoding
label_encoder = LabelEncoder()
df['Label_Encoded'] = label_encoder.fit_transform(df['Category'])
```

```python
df['Label_Encoded']
```

```python
#One-hot encoding
onehot_encoded = pd.get_dummies(df['Category'], prefix='Category')
df = pd.concat([df, onehot_encoded], axis=1)
```

#**2. Creating Derived columns**

```python
from datetime import datetime
```

```python
#Example data
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Date_of_Birth': ['1990-05-14', '1985-08-22', '1995-02-10']}
df = pd.DataFrame(data)
```

```python
type(data)
```

```python
df
```

```python
df.info()
```

```python
type(df)
```

```python
#Convert Date_of_Birth to datetime
df['Date_of_Birth'] = pd.to_datetime(df['Date_of_Birth'])
```

```python
df['Date_of_Birth']
```

```python
#Calculating Age
current_year = datetime.now().year
df['Age'] = current_year - df['Date_of_Birth'].dt.year
```

```python
df['Age']
```

#**3. Grouping and Aggregating Data**

```python
#Example data
data = {'Category': ['A', 'A', 'B', 'B', 'C'],
        'Value': [10, 20, 10, 30, 40]}
df = pd.DataFrame(data)
```

```python
type(df)
```

```python
type(data)
```

```python
df
```

```python
#Grouping by aggregate
grouped_data = df.groupby('Category').agg({'Value': ['sum', 'mean']})
print("\nGrouped Data:\n")
grouped_data
```

```python
#Pivot table
pivot_table = pd.pivot_table(df, values='Value', index='Category', aggfunc=['sum', 'mean'])
print("\nPivot Table:\n")
pivot_table
```
