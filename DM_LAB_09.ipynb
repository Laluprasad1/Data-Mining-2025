## **1. Employee Salary Analysis**<br>
A company wants to analyze its employeesâ€™ salaries to determine increments based on performance. The HR department has collected the following dataset:

```python
data1 = {"Employee": ['John', 'Sarah', 'Mike', 'Emma', 'Tom'],
        "Department": ['IT', 'HR', 'IT', 'Finance', 'HR'],
        "Salary": [60000, 50000, 75000, 65000, 48000],
        "Experience(Years)": [5, 3, 7, 4, 2],
        "Performance_Rating(Out of 5)": [4.5, 3.8, 4.9, 4.2, 3.5],
        }
```

**1.	Create a DataFrame using the given data.**

```python
df = pd.DataFrame(data1)
```

```python
df
```

**2.	Calculate the average salary of employees in each department.**

```python
avg_salary_by_dpt = df.groupby('Department')['Salary'].mean()
avg_salary_by_dpt
```

**3.	Filter out employees who have a performance rating above 4.0.**

```python
high_performers = df[df['Performance_Rating(Out of 5)'] > 4.0]
high_performers
```

**4.	Add a new column called "Increment", where employees with a rating above 4 get a 10% increment, and others get a 5% increment.**

```python
df['Increment'] = np.where(df['Performance_Rating(Out of 5)'] > 4, 0.10, 0.05)
```

```python
df
```

```python
# Calculate new salary after increment
df['new_salary'] = df['Salary'] * (1 + df['Increment'])
```

```python
df['new_salary']
```

```python
df
```

**5.	Sort employees in descending order of salary after the increment.**

```python
df_sorted = df.sort_values('new_salary', ascending = False)
```

```python
df_sorted
```

## **2. COVID-19 Cases Analysis** <br>
 A health department is monitoring COVID-19 cases in different cities. They have recorded the following data:

```python
data2 = {"City": ["New York", "Los Angeles", "Chicago", "Houston", "Miami"],
         "Total Cases": [100000, 85000, 65000, 70000, 55000],
         "Recovered": [75000, 70000, 50000, 60000, 45000],
         "Deaths": [5000, 3500, 2500, 2000, 1500]
         }
```

**1.	Create a DataFrame for the given data.**

```python
df2 = pd.DataFrame(data2)
```

```python
df2
```

**2.	Calculate the recovery rate ((Recovered / Total Cases) * 100) for each city and add it as a new column.**

```python
df2['recovery_rate'] = (df2['Recovered'] / df2['Total Cases']) * 100
df2
```

**3.	Identify cities where the death rate ((Deaths / Total Cases) * 100) is higher than 5%.**

```python
df2['death_rate'] = (df2['Deaths'] / df2['Total Cases']) * 100
df2
```

```python
high_death_rate = df2[df2['death_rate'] >= 5]
high_death_rate
```

**4.	Sort the cities by the highest recovery rate.**

```python
df2_sorted = df2.sort_values('recovery_rate', ascending=False)
df2_sorted
```

**5.	Find the total number of cases and deaths across all cities.**

```python
total_cases = df2['Total Cases'].sum()
total_deaths = df2['Deaths'].sum()
print("Total number of cases:", total_cases)
print("Total numbver of deaths:", total_deaths)
```

# **3. Online Shopping Sales Analysis** <br>
An e-commerce company wants to analyze its sales data. The dataset contains the following details:


```python
data3 = {
    'Order ID': [1001, 1002, 1003, 1004, 1005],
    'Customer': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop', 'Phone'],
    'Price': [800, 500, 300, 750, 600],
    'Quantity': [1, 2, 1, 1, 1],
    'Discount (%)': [10, 5, 15, 8, 12]
}
```

**1.	Create a DataFrame from the given sales data.**

```python
df3 = pd.DataFrame(data3)
df3
```

**2.	Calculate the total price per order after applying the discount. (Total Price = (Price * Quantity) - Discount)**

```python
df3['Total price'] = (df3['Price'] * df3['Quantity']) - (df3['Price'] * df3['Quantity'] * df3['Discount (%)'] / 100)
df3
```
