#**Simple Linear Regression**

The equation of a simple linear regression model is:<br>
Y=mX+b+ε<br>
Where:<br>
• Y = Dependent variable (Target)<br>
• X = Independent variable (Feature)<br>
• m= Slope of the line (Coefficient)<br>
• b = Intercept (where the line crosses the Y-axis)<br>
• ε = Error term (Random noise in data)<br>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = 2*np.random.rand(100, 1)
y = 4 + 3*x + np.random.randn(100, 1)
```

```python
plt.scatter(x,y, color='blue')
plt.xlabel('Independent Variable(x)')
plt.ylabel('Dependent Variable(y)')
plt.title('Simple Linear Regression')
plt.show()
```

**Splitting Data into training nad Testing**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

**Training the Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

m = model.coef_       #Slope
b = model.intercept_    #Intercept
```

```python
print(f"Slope(m): {m}")
print(f"Intercept(b): {b}")
```

**Making Predictions**

```python
y_pred = model.predict(x_test)
```

**Visualizing the Regression Line**

```python
plt.scatter(x, y, color="blue", label="Actual Data")
plt.plot(x, model.predict(x), color="red", linewidth=2, label="Regression Line")
plt.xlabel("Independent Variable (x)")
plt.ylabel("Dependent Variable (y)")
plt.legend()
plt.title("Linear Regression Fit")
plt.show()
```

**Evaluating the Model**

```python
from sklearn.metrics import mean_squared_error
```

```python
#calculate MSE
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```
